### Erlang C Formula


```python
import numpy as np
```

```python
E = 2800
m = 800
```

```python
def Pw(E, m):
    denom1 = 0
    for i in range(m):
        headache = (i - m) * np.log(E) - 0.5 * np.log(2 * np.pi * i) - i * np.log(i / np.e)
        print(np.exp(headache))
        denom1 += np.exp(headache)
    numerator = m / (m - E)
    pw = numerator / (denom1 + numerator)
    return pw
```

```python
print(Pw(E, m))
```
