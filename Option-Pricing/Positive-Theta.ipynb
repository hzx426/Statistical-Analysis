Options usually have negative thetas, but close to maturity deep ITM put options may have positive thetas due to the interest rate effect -- exercising earlier discounts the strike less.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

import pricer.formulas as pricer
```

```python
timeToMats = 0.1 * np.arange(1, 80)
thetas = []
myOptionObject = pricer.Option()
for timeToMat in timeToMats:
    greeks = myOptionObject.bsGreeks(parity='put', spot=12, strike=15, vol=0.30, intRate=0.02, divYield=0.0, timeToMat=float(timeToMat))
    thetas.append(greeks['Theta'])

plt.plot(timeToMats, thetas)
plt.show()
```
